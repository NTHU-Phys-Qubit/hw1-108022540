```python
import numpy as np
import matplotlib.pyplot as plt
```

# 2D Random walk

We will concider three kinds of 2D random talk.
* $dx$ and $dy$ are uniformly distributed between $[-1/2, +1/2]$.
* $dx=+1$ with probability 1/2 and $dx=-1$ with probability 1/2. Similarly for $dy$.
* $(dx, dy)=(\cos(\theta), \sin(\theta))$ where $\theta$ is uniformly distributed between $[0, 2\pi]$.

For all kinds of the 2D random walk, you should do the following:
* Plot a single trajectory.
* Plot the distribution of the end point for an ensemble of trajectories.
* Plot the histogram of the $r=x^2_f+y^2_f$ for an ensemble of trajectories.

```python
x = 0
y = 0
py = [y]
px = [x]
for i in range (100):
    dy = np.random.random()-0.5
    dx = np.random.random()-0.5
    x = x + dx
    y = y + dy
    py.append(y)
    px.append(x)
plt.figure(figsize=(10,5))
plt.plot(px,py)
plt.title("2D random walk")
plt.xlabel("x position")
plt.ylabel("y position")


x = 0
px = [x]
y = 0
py = [y]
for I in range(800):   
    for i in range (800):
        x = 0
        dx = np.random.random()-0.5
        x = x + dx
    px.append(x)

    for i in range (800):
        y = 0
        dy = np.random.random()-0.5
        y = y + dy
    py.append(y)
plt.figure(figsize=(10,5))
plt.plot(px,py,'o')
plt.xlabel("end point x position")
plt.ylabel("end point y position")


x = 0
px = [x]
y = 0
py = [y]
for I in range(2000):
    for i in range (2000):
        x = 0
        dx = np.random.random()-0.5
        x = x + dx
    px.append(x)
    pxa = np.array(px)

    for i in range (2000):
        y = 0
        dy = np.random.random()-0.5
        y = y + dy
    py.append(y)
    pya = np.array(py)
r = ((pxa)**2 + (pya)**2)**0.5
plt.figure(figsize=(10,5))
plt.hist(r,bins=50)
plt.xlabel("end point distance (r)")
plt.ylabel("times")
```
